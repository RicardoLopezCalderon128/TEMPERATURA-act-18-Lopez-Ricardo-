## 1. Fórmulas de Conversión

Para convertir temperaturas, nos basamos en la relación lineal entre las escalas Celsius y Fahrenheit.

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(También se puede usar $1.8$ en lugar de $9/5$)*

* **De Fahrenheit a Celsius:**
    $$°C = \frac{(°F - 32) \times 5}{9}$$
    *(O dividir entre $1.8$ después de restar $32$)*



---

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir 25°C a Fahrenheit
1.  **Identificar datos:** Tenemos $°C = 25$.
2.  **Sustituir en la fórmula:** $°F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar:** $45 + 32 = 77$.
* **Resultado:** 25°C equivalen a **77°F**.

### Ejemplo B: Convertir 95°F a Celsius
1.  **Identificar datos:** Tenemos $°F = 95$.
2.  **Sustituir en la fórmula:** $°C = (95 - 32) \times 5/9$.
3.  **Restar:** $95 - 32 = 63$.
4.  **Multiplicar y dividir:** $(63 \times 5) / 9 = 315 / 9 = 35$.
* **Resultado:** 95°F equivalen a **35°C**.

## 3. Código en Python (Interactivo)

Este programa utiliza funciones específicas y un bucle `while` para mantenerse en ejecución hasta que decidas salir.

```python
def convertirCF():
    celsius = float(input("\nIngrese los grados Celsius a convertir: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"---> {celsius}°C es igual a {fahrenheit:.2f}°F")

def convertirFC():
    fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"---> {fahrenheit}°F es igual a {celsius:.2f}°C")

def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES DE TEMPERATURA ---")
        print("1. Convertir grados Celsius a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Celsius")
        print("3. Salir")
        
        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida, por favor intente de nuevo.")

# Ejecución del programa
if __name__ == "__main__":
    menu()
```

### Explicación del Código:
* **`float()`**: Permite que el usuario ingrese números con decimales (ej. 36.5).
* **`f-strings`**: Usamos `f"{variable:.2f}"` para mostrar solo 2 decimales y que el resultado se vea limpio.
* **`while True`**: Crea un ciclo infinito que solo se rompe (`break`) cuando el usuario elige la opción 3.
* **`if/elif/else`**: Controla el flujo del programa según la entrada del usuario.
